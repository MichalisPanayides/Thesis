```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
server_1 = np.array([
    [1,      1,      1.5, 1.5],
    [np.NaN, np.NaN, 1.5, 1.5],
])
server_2 = np.array([
    [0.8,    0.8,    0.8, 2],
    [np.NaN, np.NaN, 0.8, 2],
])
```

```python
plt.figure(figsize=(8, 3))
plt.imshow(server_1)
plt.xticks(np.arange(0, 4, 1))
plt.yticks(np.arange(0, 2, 1))
plt.xlabel('v')
plt.ylabel('u')
plt.title("Server 1 service rate for all states (u, v)")
plt.clim(0, 2)
plt.colorbar()
plt.axhline(0.5, color='black', linestyle='dotted')
plt.axvline(0.5, color='black', linestyle='dotted')
plt.axvline(1.5, color='black', linestyle='dotted')
plt.axvline(2.5, color='black', linestyle='dotted')
plt.savefig('server_1_1.pdf', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8, 3))
plt.imshow(server_2)
plt.xticks(np.arange(0, 4, 1))
plt.yticks(np.arange(0, 2, 1))
plt.xlabel('v')
plt.ylabel('u')
plt.title("Server 2 service rate for all states (u, v)")
plt.clim(0, 2)
plt.colorbar()
plt.axhline(0.5, color='black', linestyle='dotted')
plt.axvline(0.5, color='black', linestyle='dotted')
plt.axvline(1.5, color='black', linestyle='dotted')
plt.axvline(2.5, color='black', linestyle='dotted')
plt.savefig('server_2_1.pdf', bbox_inches='tight')
plt.show()
```
